```python
import os
import openslide
from PIL import Image

# Function to generate and save thumbnail of NDPI file
def generate_thumbnail(svs_file, output_folder):
    # Open NDPI file
    slide = openslide.OpenSlide(svs_file)
    
    # Get thumbnail image
    thumbnail = slide.get_thumbnail((1024, 1024))
    
    # Save thumbnail with the same filename in the output folder
    thumbnail_path = os.path.join(output_folder, os.path.splitext(os.path.basename(svs_file))[0] + '.png')
    thumbnail.save(thumbnail_path, format='PNG')

# Path to the folder containing NDPI files
input_folder = 'SVS'

# Path to the folder where thumbnails will be saved
output_folder = 'thumbnails'

# Create the output folder if it doesn't exist
os.makedirs(output_folder, exist_ok=True)

# Iterate through NDPI files in the input folder and generate thumbnails
for filename in os.listdir(input_folder):
    print(filename)
    #if filename.endswith('.ndpi'):
    svs_file = os.path.join(input_folder, filename)
    print(svs_file)
    generate_thumbnail(svs_file, output_folder)
```
