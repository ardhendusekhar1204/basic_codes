```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.colors as mcolors

# Define the colormap: blue for minimum, spectrum of colors in between, red for maximum
colors = ['blue', 'cyan', 'green', 'yellow', 'orange', 'red']  # All colors in between
n_bins = 256  # Number of bins for interpolation (higher value for smoother gradient)
cmap_name = 'blue_to_red_spectrum'

# Create the colormap
cmap = mcolors.LinearSegmentedColormap.from_list(cmap_name, colors, N=n_bins)

# Generate some data to demonstrate the colormap
data = np.random.rand(10000, 10000)  # Replace with your actual data

# Plot using the custom colormap
plt.imshow(data, cmap=cmap)
plt.colorbar()  # Show the color scale
plt.title('Custom Colormap: Blue to Red Spectrum')
plt.show()
```
